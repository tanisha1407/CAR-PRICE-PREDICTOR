```python
#adding an age column in the dataset
```

```python
import datetime
```

```python
date_time=datetime.datetime.now()
```

```python
data['Age']=date_time.year-data['Year']
```

```python
data.head()
```

```python
data.drop('Year',axis=1,inplace=True)#dropping year column
```

```python
data.head()
```

```python
#removing outliers:
import seaborn as sns
```

```python
sns.boxplot(data['Selling_Price'])
```

```python
#considering the two points b/w 33 and 35 as outliers
sorted(data['Selling_Price'],reverse=True)
```

```python
data=data[~(data['Selling_Price']>=33.0) & (data['Selling_Price']<=35.0)] #used ~ for removing outliers
```

```python
data.shape
```

```python
#Encoding the Categorical Columns
```

```python
data.head(1)
```

```python
#categorical columns= fuel_type,seller_type,transmision
```

```python
data['Fuel_Type'].unique()
```

```python
data['Fuel_Type']=data['Fuel_Type'].map({'Petrol':0, 'Diesel':1, 'CNG':2})
```

```python
data['Seller_Type'].unique()
```

```python
data['Seller_Type']=data['Seller_Type'].map({'Dealer':0, 'Individual':1})
```

```python
data['Transmission'].unique()
```

```python
data['Transmission'] =data['Transmission'].map({'Manual':0,'Automatic':1})
```

```python
data.head()
```

```python
# Store Feature Matrix In X and Response(Target) In Vector y
```

```python
X = data.drop(['Car_Name','Selling_Price'],axis=1)
y = data['Selling_Price']
```

```python
y
```

```python
#splitting dataset:
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
#models:

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
#model training:
lr = LinearRegression()
lr.fit(X_train,y_train)

rf = RandomForestRegressor()
rf.fit(X_train,y_train)

xgb = GradientBoostingRegressor()
xgb.fit(X_train,y_train)

xg = XGBRegressor()
xg.fit(X_train,y_train)
```

```python
# Prediction on Test Data
```

```python
y_pred1 = lr.predict(X_test)
y_pred2 = rf.predict(X_test)
y_pred3 = xgb.predict(X_test)
y_pred4 = xg.predict(X_test)
```

```python
#evaluation:
from sklearn import metrics
score1 = metrics.r2_score(y_test,y_pred1)
score2 = metrics.r2_score(y_test,y_pred2)
score3 = metrics.r2_score(y_test,y_pred3)
score4 = metrics.r2_score(y_test,y_pred4)
print(score1,score2,score3,score4)
```

```python
final_data = pd.DataFrame({'Models':['LR','RF','GBR','XG'],
             "R2_SCORE":[score1,score2,score3,score4]})
```

```python
final_data
```

```python
sns.barplot(data=final_data, x='Models', y='R2_SCORE')
```

```python
#saving the model:
xg = XGBRegressor()
xg_final = xg.fit(X,y)
```
